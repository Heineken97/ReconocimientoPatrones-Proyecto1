```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost_function(X, y, w,b):
    m = len(y)
    L = np.sum(((X.dot(w)+b) - y) ** 2) / (2 * m)
    return L

# Definir la funci√≥n de descenso de gradiente
def gradient_descent(X, y, alpha, num_iters):
    m, n = X.shape
    L_history = np.zeros(num_iters)
    W_history= np.zeros(num_iters)
    B_history= []
    b=5
    w = np.zeros((n, 1))

    for i in range(num_iters):
        w = w - (2*alpha / m) * X.T.dot((X.dot(w)+b) - y)
        b= b -(2*alpha / m) * ((X.dot(w)+b) )
        W_history[i]=w
        L_history[i] = cost_function(X, y, w,b)

    return w, L_history,W_history, B_history
```

```python
# Generar datos de ejemplo
X = 2 * np.random.rand(1000, 1)
y = 4 + 3 * X + np.random.randn(1000, 1)


# Normalizar los datos
X = (X - np.mean(X)) / np.std(X)
y = (y - np.mean(y)) / np.std(y)

# A√±adir la columna de 1s a la matriz X
X_b = np.c_[np.ones((len(X), 1)), X]


# Definir la tasa de aprendizaje y el n√∫mero de iteraciones
alpha = 0.591
num_iters = 10

# Ejecutar el algoritmo de descenso de gradiente
w, J_history,W_history,B_history = gradient_descent(X, y, alpha, num_iters)

# Imprimir los par√°metros √≥ptimos del modelo
print('w:', w)

# Trazar la curva de aprendizaje
plt.plot(J_history)
plt.title('Loss')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()
```
